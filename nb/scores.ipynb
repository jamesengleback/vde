```python
def softmax(arr):
    e = np.exp(arr - max(arr))
    return e / sum(e)
```

```python
def synth_res():
    affs = np.random.uniform(-2,-10,size=(8))
    dists = np.random.uniform(5, 20,size=(8))
    return affs, dists

synth_res()
```

```python
def score_a(affinities, distances):
    return np.mean(distances) - np.log(abs(affinities).mean()), \
           distances.mean(), \
           affinities.mean()

def score_b(affinities, distances):
    return np.mean(distances * softmax(abs(affinities))), \
           distances.mean(), \
           affinities.mean()

def score_c(affinities, distances):
    softAff = softmax(abs(affinities))
    e = sum(affinities)
    score = np.mean(softAff * distances)  - np.log(abs(e))
    return score, \
           distances.mean(), \
           affinities.mean()

score_c(*synth_res())
```

```python
def viz(sfxn):
    data = [synth_res() for _ in range(1000)]
    xyz = np.array([sfxn(*i) for i in data])
    plt.scatter(xyz[:,1],xyz[:,2],c=xyz[:,0],s=2)
    plt.xlabel('dist')
    plt.ylabel('aff')
    plt.colorbar(label='score')
    plt.show()

viz(score_c)
```
